```python
import librosa as lr 
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
audio, sfreq= lr.load('G:\project\open-the-goddamn-door.wav')
time= np.arange(0,len(audio)) / sfreq

time
len(audio)
```

```python
fig, ax = plt.subplots()
ax.plot(time,audio)
ax.set(xlabel='Time',ylabel='Sound Amplitude')
plt.show()
ipd.Audio(audio, rate=sfreq)
```

```python
time= np.arange(0,len(audio)) / sfreq*2
fig, ax = plt.subplots()
ax.plot(time,audio)
ax.set(xlabel='Time',ylabel='Sound Amplitude')
plt.show()
ipd.Audio(audio, rate=sfreq*2)
```

```python
# pitch shifting
audio_shift=lr.effects.pitch_shift(audio, sfreq, 2.5, 12)
time2=np.arange(0,len(audio_shift)) / sfreq
fig, ax = plt.subplots()
ax.plot(time2,audio_shift)
ax.set(xlabel='Time',ylabel='Sound Amplitude')
plt.show()
ipd.Audio(audio_shift, rate=sfreq)
```

```python
audio_fast=lr.effects.time_stretch(audio, 2)
time3=np.arange(0,len(audio_fast)) / sfreq
fig, ax = plt.subplots()
ax.plot(time3,audio_fast)
ax.set(xlabel='Time', ylabel = 'Amplitude')
plt.show()
ipd.Audio(audio_fast, rate=sfreq)
```

```python
audio_slow=lr.effects.time_stretch(audio, 0.5)
time4=np.arange(0,len(audio_slow)) / sfreq
fig, ax = plt.subplots()
ax.plot(time4,audio_slow)
ax.set(xlabel='Time', ylabel = 'Amplitude')
plt.show()
ipd.Audio(audio_slow, rate=sfreq)
```
